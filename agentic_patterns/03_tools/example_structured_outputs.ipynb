```python
import rich
from aixtools.agents import get_agent, run_agent
```

## Example: Structured outputs

```python
from pydantic import BaseModel, Field

class ProgrammingLanguage(BaseModel):
    name: str
    paradigm: str
    popularity_percent: float = Field(..., description="Popularity of the programming language in percent")
```

```python
agent = get_agent(output_type=list[ProgrammingLanguage]) # type: ignore

prompt = "What are the top 3 most popular programming languages?"
ret, nodes = await run_agent(agent, prompt, verbose=True, debug=True, log_model_requests=True)
```

```python
# Show the result
rich.print(ret)
```
