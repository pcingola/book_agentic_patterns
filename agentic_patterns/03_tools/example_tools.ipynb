# Tools

## Initialize

```python
import rich
from aixtools.agents import get_agent, run_agent
```

### Create tools

```python
def add(a: int, b: int) -> int:
    """Add two numbers"""
    print(f"Adding {a} + {b}")
    return a + b

def sub(a: int, b: int) -> int:
    """Substract two numbers"""
    print(f"Substracting {a} - {b}")
    return a - b
```

### Use tools in agent

```python
agent = get_agent(tools=[add, sub])

prompt = "What is the sum of 40123456789 and 2123456789?"
# ret = await run_agent(agent, prompt, verbose=True, debug=True, log_model_requests=True)
ret, nodes = await run_agent(agent, prompt, log_model_requests=True, verbose=True, debug=True)
print(ret)
```
