````python
import uuid
import rich
from fasta2a.client import A2AClient
from fasta2a.schema import Message, TextPart
````

## A2A Client

````python
client = A2AClient(base_url="http://127.0.0.1:8000")
````

### Create a message

````python
prompt = "What is the sum of 40123456789 and 2123456789?"

text_part = TextPart(kind="text", text=prompt)

parts = list()
parts.append(text_part)

message = Message(kind="message", role="user", parts=parts, message_id=str(uuid.uuid4()))
````

### Send a task

````python
ret = await client.send_message(message=message)
rich.print(ret)
task_id = ret['result']['id'] if 'result' in ret and 'id' in ret['result'] else ''
````

````python
import asyncio
from IPython.display import clear_output

SLEEP_TIME = 0.2
state = None
iteration = 0

while state != "completed":
    ret = await client.get_task(task_id=task_id)
    # Clear the cell's output
    clear_output(wait=True)
    print(f"Iteration {iteration}")
    rich.print(ret)
    # Check the state of the task
    state = ret['result']['status']['state'] if 'result' in ret and 'status' in ret['result'] else None
    if state == "completed":
        break
    elif state == "failed":
        raise Exception("Task failed")
    # Sleep for a while before checking again
    await asyncio.sleep(SLEEP_TIME)
    iteration += 1
````
