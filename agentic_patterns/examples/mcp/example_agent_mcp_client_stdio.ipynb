# Agent with MCP Tools (STDIO)

This example connects an agent to an MCP server via STDIO transport. The server exposes tools that the agent can discover and invoke.

```python
from pydantic_ai.mcp import MCPServerStdio
from agentic_patterns.core.agents import get_agent, run_agent
```

```python
# Connect to MCP server via STDIO (spawns server as subprocess)
server = MCPServerStdio(
    command="fastmcp", args=["run", "-t", "stdio", "example_mcp_server.py"]
)
```

```python
# Create agent with MCP server as toolset
agent = get_agent(toolsets=[server])
```

```python
# Run agent - it discovers and uses the 'add' tool from the MCP server
async with agent:
    result, nodes = await run_agent(
        agent, "What is 40123456789 plus 2123456789?", verbose=True
    )
```

```python
print(f"Answer: {result.result.output}")
```
