# Simple LLM: Translation (System Prompt)

```python
import rich
from agentic_patterns.core.agents import get_agent, run_agent
```

```python
language = "French"

# System prompt
system_prompt = f"Translate into {language}"
agent = get_agent(system_prompt=system_prompt)

# User prompt
prompt = "I like coffee."

# Run agent with verbose output
ret, nodes = await run_agent(agent, prompt, verbose=True)
rich.print(ret)
```
