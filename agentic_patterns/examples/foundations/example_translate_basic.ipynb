# Simple LLM: Translation (Basic)

```python
import rich
from agentic_patterns.core.agents import get_agent, run_agent
```

```python
# Get agent.
# The function 'get_agent()' is a lightweight wrapper function to get Pydantic-AI's Agent object.
# It uses the parameters defined in config.yaml file ('model' section), if no 'config_name' is provided,
# it uses the 'default' configuration entry.
agent = get_agent()
```

```python
prompt = "Translate to French: I like coffee."
```

```python
# Run agent: Runs the agent with the given prompt and returns the agent run object and the list of model messages.
agent_run, nodes = await run_agent(agent, prompt)

assert agent_run is not None and agent_run.result is not None
print(agent_run.result.output)
```

```python
# Show the output
assert agent_run is not None and agent_run.result is not None
print(agent_run.result.output)
```

```python
# Show agent run details (result and token usage)
rich.print(agent_run)
```

```python
# Show node details
for node in nodes:
    rich.print(node)
```
