# Agent with MCP (STDIO)

**IMPORTANT:** For this notebook you need to run MCP server:
```bash
./scripts/example_10_mcp_server.sh
```

````python
from pydantic_ai.mcp import MCPServerStreamableHTTP
from aixtools.agents import get_agent, run_agent
````

````python
server = MCPServerStreamableHTTP(url='http://127.0.0.1:8000/mcp/')
````

````python
agent = get_agent(toolsets=[server])
````

````python
async with agent:
    ret = await run_agent(agent, "What is the add of 40123456789 and 2123456789?", verbose=True, debug=True)
````

````python
print(ret[0])
````
