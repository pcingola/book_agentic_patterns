# Agent with MCP (STDIO)

```python
from pydantic_ai.mcp import MCPServerStdio
from aixtools.agents import get_agent, run_agent
```

```python
server = MCPServerStdio(command='fastmcp', args=['run', 'example_07_mcp_server.py'])
```

```python
agent = get_agent(toolsets=[server])
```

```python
async with agent:
    ret = await run_agent(agent, "What is the addition of 40123456789 and 2123456789?", verbose=True, debug=True)
```

```python
print(ret[0])
```
